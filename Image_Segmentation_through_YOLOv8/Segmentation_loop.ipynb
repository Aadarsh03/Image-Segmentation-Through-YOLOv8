```python
from ultralytics import YOLO 
from IPython.display import Image
import os

# Initialize YOLOv8 model
yolo8_model_segmentation = YOLO('yolov8n-seg.pt')

# Specify the input and output folders
input_folder = "input_images"
output_folder = "output_images"

# Check if the input folder exists
if not os.path.exists(input_folder):
    print(f"Error: Input folder '{input_folder}' does not exist.")
else:
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Process all images in the input folder
    for image_name in os.listdir(input_folder):
        if image_name.endswith(('.jpg', '.jpeg', '.png')):
            input_path = os.path.join(input_folder, image_name)
            output_path = os.path.join(output_folder, f"segmented_{image_name}")

            # Perform segmentation using YOLOv8
            yolo8_model_segmentation.predict(source=input_path, save_txt=True, save=True, exist_ok=True)

            # Rename the segmented image to the desired output folder
            os.rename(f"runs/segment/predict/{image_name}", output_path)

            # Display the segmented image (optional)
            display(Image(filename=output_path, height=300, width=500))
```
